```python
'''
    傅里叶变换
'''
import cv2
import numpy as np
from scipy.fftpack import fft,ifft
import matplotlib.pyplot as plt
from matplotlib.pylab import mpl
import IPython.display as ipd
import math

# 查看原图
img = cv2.imread('image/lena.jpg',0)
plt.imshow(img,cmap="gray")
```

```python
img.shape
```

```python
# 快速傅里叶变换
img_fft = np.fft.fft2(img)
# 原点移动
img_fft = np.fft.fftshift(img_fft)
# 对数变换
img_fft = np.log(1 + np.abs(img_fft))
plt.imshow(img_fft,cmap="gray")
```

```python
# 分离幅度谱和相位谱
dft = np.fft.fft2(img)
dft_shift = np.fft.fftshift(dft)
# 幅度谱
magnitude_spectrum = np.abs(dft_shift)
# 相位谱
phase_spectrum = np.angle(dft_shift)
```

```python
# 复原
ff2fft = np.real(np.fft.ifft2(np.abs(dft)*np.exp(1j*(np.angle(dft)))))
plt.imshow(ff2fft,cmap="gray")
```

```python
# 相位谱为0
fft2fft = np.real(np.fft.ifft2(np.abs(dft)))
plt.imshow(ff2fft,cmap="gray")
```

```python
# 幅度谱为A
A = 2
ff2fft = np.real(np.fft.ifft2(A*np.exp(1j*np.angle(dft))))
plt.imshow(ff2fft,cmap="gray")
```
